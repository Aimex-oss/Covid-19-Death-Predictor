```python
unrelevant_columns = ["SEX","PREGNANT","COPD","ASTHMA","INMSUPR","OTHER_DISEASE","CARDIOVASCULAR",
                      "OBESITY","TOBACCO"]

df.drop(columns=unrelevant_columns,inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=["MEDICAL_UNIT","CLASIFFICATION_FINAL"],drop_first=True)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df.AGE = scaler.fit_transform(df.AGE.values.reshape(-1,1))
```

```python
x = df.drop(columns="DEATH")
y = df["DEATH"]
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.2, random_state=42)
print("Train_x :",train_x.shape)
print("Test_x :",test_x.shape)
print("Train_y :",train_y.shape)
print("Test_y :",test_y.shape)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(train_x,train_y)
print("Logistic Regression Accuracy :",logreg.score(test_x, test_y))
```

```python
from sklearn.metrics import f1_score

print("Logistic Regression F1 Score :",f1_score(test_y,logreg.predict(test_x),average=None))
```

```python
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(test_y, logreg.predict(test_x)), annot=True, fmt=".0f")
plt.title("Logistic Regression Confusion Matrix",fontsize=18, color="red");
```

```python
ax = sns.countplot(df.DEATH, palette=sns.cubehelix_palette(2))
plt.bar_label(ax.containers[0])
plt.title("Death Distribution", fontsize=18,color="red");
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=0)
x_resampled,y_resampled = rus.fit_resample(x,y)
```

```python
ax = sns.countplot(y_resampled, palette=sns.cubehelix_palette(2))
plt.bar_label(ax.containers[0])
plt.title("Death Distribution", fontsize=18,color="red");
```

```python
train_x, test_x, train_y, test_y = train_test_split(x_resampled,y_resampled, test_size=0.2, random_state=42)
print("Train_x :",train_x.shape)
print("Test_x :",test_x.shape)
print("Train_y :",train_y.shape)
print("Test_y :",test_y.shape)
```

```python
logreg = LogisticRegression()
logreg.fit(train_x,train_y)
print("Logistic Regression Accuracy :",logreg.score(test_x, test_y))
```

```python
print("Logistic Regression F1 Score :",f1_score(test_y,logreg.predict(test_x),average=None))
```

```python
sns.heatmap(confusion_matrix(test_y, logreg.predict(test_x)), annot=True, fmt=".0f")
plt.title("Logistic Regression Confusion Matrix",fontsize=18, color="red");
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(train_x, train_y)
```

```python
from sklearn.metrics import accuracy_score

print('Random Forest Accuracy Score :', accuracy_score(test_y,random_forest.predict(test_x)))
```

```python
sns.heatmap(confusion_matrix(test_y, random_forest.predict(test_x)), annot=True, fmt='.0f')
plt.title("Random Forest Confusion Matrix", fontsize=18)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix
import matplotlib.pyplot as plt
clf = DecisionTreeClassifier()
clf.fit(train_x, train_y)
```

```python
y_pred = clf.predict(test_x)
```

```python
accuracy = accuracy_score(test_y, y_pred)
print("Accuracy:", accuracy)
```

```python
confusion = confusion_matrix(test_y, y_pred)
plot_confusion_matrix(clf, test_x, test_y)
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20,6))
print("Confusion Matrix of our Models using under Sampling")
sns.heatmap(confusion_matrix(test_y, logreg.predict(test_x)), annot=True, fmt=".0f", ax=ax[0])
ax[0].set_title("Logistic Regression Confusion Matrix", fontsize=18)
print("Logistic Regression Accuracy:",logreg.score(test_x, test_y))

sns.heatmap(confusion_matrix(test_y, random_forest.predict(test_x)), annot=True, fmt=".0f", ax=ax[1])
ax[1].set_title("Random Forest Confusion Matrix", fontsize=18)
print('Random Forest Accuracy:', accuracy_score(test_y,random_forest.predict(test_x)))

sns.heatmap(confusion_matrix(test_y, clf.predict(test_x)), annot=True, fmt='.0f', ax=ax[2])
ax[2].set_title("Decision Tree Confusion Matrix", fontsize=18)
print("Decision Tree Accuracy:", accuracy)

plt.show()
```
